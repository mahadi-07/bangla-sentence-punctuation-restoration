```python
from transformers import AutoTokenizer
from datasets import load_dataset

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
dataset = load_dataset("cornell-movie-review-data/rotten_tomatoes", split="train")
```

```python
dataset
```

```python
tokenizer(dataset[0]['text'])
```

```python
def tokenization(example):
    return tokenizer(example["text"])

data = dataset.map(tokenization, batched=True)
data
```

```python
data.column_names
```

```python
data.set_format(type="torch", columns=["input_ids", "token_type_ids", "attention_mask",  "label"])
```

```python
from transformers import AutoFeatureExtractor, Processor
from datasets import load_dataset, Audio
```

```python
feature_extractor = AutoFeatureExtractor.from_pretrained("facebook/wav2vec2-base-960h")
dataset = load_dataset("PolyAI/minds14", "en-US", split="train")
```

```python
audio = dataset[0]["audio"]
```

```python
audio.get_all_samples().sample_rate
```

```python
dataset = dataset.cast_column("audio", Audio(sampling_rate=16_000))
audio = dataset[0]["audio"]
audio
```

```python
def preprocess_function(examples):
    audio_arrays = [x.get_all_samples().data for x in examples["audio"]]
    inputs = processor(
        audio_arrays,
        sampling_rate=feature_extractor.sampling_rate,
        max_length=16000,
        truncation=True
    )
dataset = dataset.map(preprocess_function, batched=True)
```
